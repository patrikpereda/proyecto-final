### Ejercicio 1: Lea la ganancia total de todos los meses y muéstrela mediante un [gráfico de líneas](https://matplotlib.org/stable/plot_types/basic/plot.html#sphx-glr-plot-types-basic-plot-py).
Se proporcionan datos de ganancia total para cada mes. El gráfico de líneas generado debe incluir las siguientes propiedades:


- X label name = Month Number
- Y label name = Total profit

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')

# ReadingData
raw_data = pd.read_csv("../data/company_sales_data.csv")
raw_data
```

image.png

> Hint: plt.plot

```python
ax = plt.axes()
ax.plot(raw_data["month_number"], raw_data["total_profit"], "b-")
ax.set(ylim=(0, 500000),
       xlabel='Month number', ylabel='Total profit',
       title='Company profit per month');
```

### Ejercicio 4: Lee los datos de ventas de toothpaste de cada mes y muéstralos usando un [scatter plot](https://matplotlib.org/stable/plot_types/basic/scatter_plot.html#sphx-glr-plot-types-basic-scatter-plot-py).
Además, agrega un [grid](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html) en el gráfico (el estilo de la línea de la cuadrícula debe ser “–“).

image.png

> Hint: plt.scatter

```python
fig_4 = plt.figure(figsize=(12, 8))
ax_4 = fig_4.add_subplot(1,1, 1)

ax_4.set_title("Tooth Paste Sales Data", fontsize=28)
plt.xlabel("Month number", fontsize=15)
plt.ylabel("Number of Units Sold", fontsize=15)
plt.ylim([4250, 8500])
plt.yticks(np.arange(4500, 8251, 250), fontsize=8)
plt.xticks(np.arange(0, 13, 1), fontsize=8)

ax_4.annotate(
    "Maximum", 
    xy=(10, 8250),  # Adjust this to ensure it points to the correct data
    xycoords='data',
    bbox=dict(boxstyle="round", fc="white", ec="gray"),
    xytext=(6, 7000),  # Move text slightly higher
    textcoords='data', 
    ha='center',
    arrowprops=dict(arrowstyle="->",
                    connectionstyle="angle3,angleA=0,angleB=-90",
                    color="black",
                    linewidth=2)
)

plt.scatter(raw_data["month_number"], raw_data["toothpaste"], color="blue", marker="o", label="Tooth Paste", s=150)
plt.legend(loc="upper left", facecolor="white", fontsize=8, borderpad=1)
plt.grid(True, linestyle="--", linewidth="1.5")
plt.show()
```

### Ejercicio 5: Lea los datos de ventas de los productos de "cream" y "facewash" y muéstrelos utilizando el [bar chart](https://matplotlib.org/stable/plot_types/basic/bar.html#sphx-glr-plot-types-basic-bar-py).
El gráfico de barras debe mostrar la cantidad de unidades vendidas por mes de cada producto. Agregue una barra separada para cada producto en el mismo gráfico.

image.png

> Hint: plt.bar

```python
month_list = raw_data['month_number'].tolist()
facecream_sales  = raw_data ['facecream'].tolist()
facewash_sales   = raw_data['facewash'].tolist()

bar_width = 0.25

fig_ex5 = plt.figure(figsize=(10, 10), dpi=80)
plt.bar([a-bar_width for a in month_list], facecream_sales, width = bar_width, label = 'Face Cream sales data', align = 'edge')
plt.bar(month_list, facewash_sales, width = bar_width, label = 'Face Wash sales data', align = 'edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')
plt.xticks(month_list)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Facewash and facecream sales data')
plt.show()
```

### Ejercicio 6: Calcule los datos de ventas totales del último año para cada producto y muéstrelos mediante un [Pie Chart](https://matplotlib.org/stable/plot_types/stats/pie.html#sphx-glr-plot-types-stats-pie-py).
Nota: En el gráfico circular debe mostrar la cantidad de unidades vendidas por año para cada producto en porcentaje.

El gráfico circular debe verse así:

image.png

> Hint: plt.pie

```python
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
sales_data   = [raw_data['facecream'].sum(), raw_data['facewash'].sum(), raw_data['toothpaste'].sum(), 
         raw_data['bathingsoap'].sum(), raw_data['shampoo'].sum(), raw_data['moisturizer'].sum()]

print(sales_data)

fig_ex8 = plt.figure(figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')
patches, texts, autotexts = plt.pie(sales_data, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.show()
```

### Ejercicio 7: Lee el `Bathing soap` y `facewash` de todos los meses y muéstralo usando un [subplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots).
el subplot debería verse así:

image.png

> Hint: plt.subplots

```python
month_list  = raw_data['month_number'].tolist()
bathing_soap   = raw_data['bathingsoap'].tolist()
facewash   = raw_data['facewash'].tolist()

f, axarr = plt.subplots(2, sharex=True , figsize=(15, 15)) # -->sharex= True les fuerza a las dos graficas compartir el eje X!
axarr[0].plot(month_list, bathing_soap, label = 'Bathingsoap Sales Data', color='k'\
              , marker='o',markersize = 15, linewidth=5)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(month_list, facewash, label = 'Face Wash Sales Data', color='r',\
              marker='o',markersize = 15, linewidth=5)
axarr[1].set_title('Sales data of  a facewash')
axarr[0].set_ylim([0, 15000])
axarr[1].set_ylim([0, 15000])
plt.xticks(month_list)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

### Ejercicio 8: Lea todos los datos de ventas de productos y muéstrelos utilizando el [stack plot](https://matplotlib.org/stable/plot_types/basic/stackplot.html#stackplot-x-y).
El gráfico de pila debería verse así:

image.png

> Hint: plt.stackplot

```python
# Extracting data
month_list  = raw_data['month_number'].tolist()
facecream   = raw_data['facecream'].tolist()
facewash   = raw_data['facewash'].tolist()
toothpaste = raw_data['toothpaste'].tolist()
boathingsoap   = raw_data['bathingsoap'].tolist()
shampoo   = raw_data['shampoo'].tolist()
moisturizer = raw_data['moisturizer'].tolist()

labels = ['FaceCream', 'FaceWash', 'Toothpaste', 'BathingSoap', 'Shampoo', 'Moisturizer']
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']

plt.figure(figsize=(10, 6))

# Stackplot
plt.stackplot(month_list, 
              facecream, facewash, toothpaste, 
              boathingsoap, shampoo, moisturizer, 
              labels=labels, colors=colors, alpha=0.7)

plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.title('Stacked Sales Data of Different Products')
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
